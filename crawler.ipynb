```python
import requests
import json
import re
import pandas
from bs4 import BeautifulSoup
from  datetime import datetime
url='http://api.roll.news.sina.com.cn/zt_list?channel=news&\
     cat_1=gnxw&cat_2==gdxw1||=gatxw||=zs-pl||=mtjj&\
     level==1||=2&show_ext=1&show_all=1&show_num=22&tag=1&\
     format=json&page={}&callback=newsloadercallback&_=1506743800654'
commentURL='http://comment5.news.sina.com.cn/page/info?version=1&format=js&channel=gn&newsid=comos-{}&group=&compress=0&ie=utf-8&oe=utf-8&page=1&page_size=20'
def getComments(newsurl):
    newsid=re.search('doc-i(.*).shtml',newsurl).group(1)
    res=requests.get(commentURL.format(newsid))
    jd=json.loads(res.text.lstrip('var data='))
    return jd['result']['count']['total']
def getNewsDetails(newsurl):
    res=requests.get(newsurl)
    res.encoding='utf-8'  
    soup=BeautifulSoup(res.text,'html.parser')
    results={}
    results['article']=soup.select('#artibodyTitle')[0].text
    #contentlist=soup.select('#artibody p')[1:-1]
    #contentstr=''
    #for i in contentlist:
        #contentstr+=i.text.strip()+'\n'
    #results['content']=contentstr
    timesource=soup.select('.time-source')[0].contents[0].strip()
    timeformat=datetime.strptime(timesource,'%Y年%m月%d日%H:%M')
    time=timeformat.strftime('%Y-%m-%d')
    results['time']=time
    try:
        results['editor']=soup.select('.article-editor')[0].text.lstrip('责任编辑：')
    except:
        results['editor']='无'
    results['comments']=getComments(newsurl)
    results['sourceCome']=soup.select('.time-source')[0].contents[1].text.strip()
    return results
def getPages(url,pageid):
    pageResult=[]
    url=url.format(pageid)
    res=requests.get(url)
    res.encoding='utf-8'
    jd=json.loads(res.text.strip().lstrip('newsloadercallback(').rstrip(');'))
    for page in jd['result']['data']:
        newsurl=page['url']
        pageResult.append(getNewsDetails(newsurl))
    return pageResult      
page_data=getPages(url,1)
dt=pandas.DataFrame(page_data)
dt
```
